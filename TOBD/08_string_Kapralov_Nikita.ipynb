1.2 Напишите функцию `show_info`, которая для рецепта по его `id` создает строку (в смысле объекта python) с описанием следующего вида:

```

----------

```

    
Данные для создания строки получите из файлов `recipes_sample_with_tags_ingredients.csv`, `steps_sample.xml` (__ЛР4__) и `tags_sample.csv` (__ЛР5__). 
Выведите созданную строку на экран.

````python
with open('data/steps_sample.xml',"r", encoding='utf-8') as file:
    data_steps = BeautifulSoup(file, 'xml')

recipe_with_steps = {}
for recipe in data_steps.find_all('recipe'):
    id = int(recipe.find("id").get_text())
    recipe_with_steps[id] = []
    
    for step in recipe.find_all("step"):
        text = step.get_text()
        recipe_with_steps[id].append(text)

````

````python
data_tags = pd.read_csv("data/recipes_sample_with_tags_ingredients.csv")
data_tags = data_tags.drop(['Unnamed: 0'], axis = 1)
data_tags
````

````python
def show_info(id_):
    try:
        print(f'''"{data[(data.id==id_)].name.to_string()[5::].title().strip()}"\n''')
        for i in recipe_with_steps[id_]:
            print(f"{recipe_with_steps[id_].index(i)}. {i}.")
        print("-"*10)
        tags = data_tags[(data_tags.id==id_)].tags.to_string()[5:-3].strip().split(';')
        for i in range(len(tags)-1):
            tags[i] = '#'+tags[i]
        tags_result = " ".join(map(str, tags))
        print(tags_result)
    except KeyError:
        print("No such a recipe")
              
show_info(44123)
````
